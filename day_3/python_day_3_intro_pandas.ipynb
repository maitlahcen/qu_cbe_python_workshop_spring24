## Selecting data

We can select particular rows using standard Python array slicing notation:

```python
data[100:105]
```

To select both rows and columns using integers, the iloc attribute can be used with the format `.iloc[rows, columns]`

```python
data.iloc[100:105,0:5]
```

We can also select both columns and rows based on labels or conditions using `.loc`

```python
data.loc[data['countrycode'] == 'QAT',['year','rgdpe']]
```

Let's say we don't want to carry around the whole dataset so we want to keep only some specific columns. To select columns, we can also pass a list containing the names of the desired columns:

```python
# Select desired columns
data = data[['countrycode','year','rgdpe','cgdpe','pop','pl_c']]
```

```python
# rename columns
data.columns = ['Country','Year','RGDP','NGDP','Population','Price level']
```

```python
# create two new columns
data.loc[:,'RGDP per capita'] = data['RGDP'] / data['Population']
data.loc[:,'Inflation'] = data['Price level'].pct_change(fill_method=None) * 100
```

Let's say now we don't need the column Price level after we used it to compute inflation. We can drop it from the dataframe:

```python
# drop a column
data = data.drop(['Price level'], axis=1)
```

```python
# descriptive statistics
data.describe()
```

```python
# select the data for one specific year using a Boolean condition
data[data['Year'] == 2019]
```

```python
# we can also combine conditions
data[(data['Year'] == 2019) & (data['RGDP per capita'] >= 100000)]
```

```python
# the object we obtain is also a dataframe. It's a modified copy (a slice) of the original dataframe. We can assign it to a variable
temp = data[data['Year'] == 2019]
type(temp)
```

```python
temp['RGDP per capita'].median()
```

## Merging data

We can add for each country its income group based on World Bank classification. This list is available in different file:

```python
# read data
temp = pd.read_csv('income_groups.csv', sep=',',header=0)
temp.head()
```

```python
# Get the number of countries in PWT
len(data['Country'].unique())
```

We will choose a `left` merge:

```python
data = pd.merge(left=data, right=temp['Income group'], how='left', left_on=['Country'], right_on=temp['Code'])
```

```python
data
```

```python
# We can plot the average rate of inflation per income group for the last available data for each country
data_last =  data.groupby('Country').last()
```

```python
data_last.groupby(['Income group'])['Inflation'].mean().plot(kind='bar', title='Average inflation rate', ylabel='%', xlabel='Income group', figsize=(6, 5))
```

We would like to calculate the YTD return of the following stocks:

```python
ticker_list = {'NVDA': 'Nvidia',
               'MSFT': 'Microsoft',
               'META': 'Meta',
               'LLY': 'Eli Lilly',
               'TM': 'Toyota',
               'AAPL': 'Apple',
               'AMZN': 'Amazon',
               'UBER': 'Uber',
               'NFLX': 'Netflix',
               'KO': 'Coca-Cola',
               'GOOG': 'Google'}
```

We will write a function that reads the list of tickers, start and end date and returns a dataframe with the data

```python
def read_data(ticker_list,start,end):

    ticker = pd.DataFrame()  # create an empty dataframe

    for tick in ticker_list:  # iterate over the list of tickers
        stock = yf.Ticker(tick)  # create an instance of `Ticker` using the ticker of the stock
        prices = stock.history(start=start, end=end)  # use the history method of yf.Ticker to get the stock price

        # Change the index to date-only
        prices.index = pd.to_datetime(prices.index.date)
        
        closing_prices = prices['Close']  # select the closing price
        
        ticker[tick] = closing_prices  # add price to the dataframe

    return ticker
```

```python
start = dt.datetime(2024, 1, 1)
end = dt.datetime(2024, 2, 25)
price_data = read_data(ticker_list,start,end)  # don't run many times to avoid being blocked by YF servers
```

```python
price_data
```

We will use the obtained data to compute the YTD return with two different methods:

```python
p1 = price_data.iloc[0]    #Get the first set of prices as a Series
p2 = price_data.iloc[-1]   #Get the last set of prices as a Series
price_change = (p2 - p1) / p1 * 100
price_change
```

```python
change = price_data.pct_change(periods=len(price_data)-1, axis='rows')*100
price_change = change.iloc[-1]
price_change
```

```python
price_change = price_change.sort_values()
price_change = price_change.rename(index=ticker_list)  # use a dictionary to rename the columns
fig, ax = plt.subplots()
ax.set_xlabel('Stock', fontsize=12)
ax.set_ylabel('YTD return in %', fontsize=12)
price_change.plot(kind='bar', ax=ax)
plt.show()
plt.close()
```
